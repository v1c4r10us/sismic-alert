```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import time

chrome_options = Options()
chrome_options.add_argument("--headless")
# Configurar el controlador del navegador
driver = webdriver.Firefox(options=chrome_options)  # Cambia a webdriver.Firefox() si prefieres Firefox

# Navegar a la página web
driver.get("https://www.data.jma.go.jp/multi/quake/index.html?lang=es") #Url de la pagina de japon
time.sleep(10)

# Obtener el contenido del DOM
dom = driver.page_source

# Imprimir el DOM
print(dom)

# Cerrar el navegador
driver.quit()
```
